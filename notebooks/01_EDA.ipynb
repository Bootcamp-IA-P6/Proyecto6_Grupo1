```python
import pandas as pd

# Cargar el Dataset
df = pd.read_csv('../data/raw/df_arabica_clean.csv')

# Visualizar como las propiedades del dataset, cuántas filas y columnas hay
print(f"Dimensiones del dataset: {df.shape}")
df.info()
```

```python
# 1. Definimos la variable objetivo (Target)
# Según la SCA, un café de especialidad es aquel que tiene 80 puntos o más.
df['is_specialty'] = (df['Total Cup Points'] >= 82.5).astype(int)

# 2. Seleccionamos las columnas que realmente aportan información al sabor y origen
# Hemos elegido estas por ser las más estándar en la industria.
cols_interes = [
    'Country of Origin', 'Arrived', 'Color', 'Aroma', 'Flavor', 
    'Aftertaste', 'Acidity', 'Body', 'Balance', 'Uniformity', 
    'Clean Cup', 'Sweetness', 'is_specialty'
]

# Creamos un nuevo dataframe solo con lo que nos importa
df_filt = df[cols_interes].copy()

# 3. Comprobamos si hay valores nulos en estas columnas
print("Valores nulos por columna:")
print(df_filt.isnull().sum())

# 4. Vemos cómo ha quedado la distribución de nuestro objetivo
print("\nDistribución de clases (0 = Comercial, 1 = Especialidad):")
print(df_filt['is_specialty'].value_counts())

display(df_filt.head())
```
