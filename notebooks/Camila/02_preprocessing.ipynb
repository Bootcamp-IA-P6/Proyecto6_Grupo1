```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# Cargamos el archivo que generamos al final del EDA
df = pd.read_csv('../../data/processed/Camila/coffee_combined_cleaned.csv')

print(f"✅ Dataset cargado con {df.shape[0]} filas.")
```

```python
import pandas as pd
import numpy as np

# 1. Identificamos qué columnas son numéricas y cuáles no
# Esto es más seguro que preguntar por 'object'
cols_numericas = df.select_dtypes(include=[np.number]).columns
cols_categoricas = df.select_dtypes(exclude=[np.number]).columns

# 2. Rellenamos las numéricas con la MEDIANA
for col in cols_numericas:
    df[col] = df[col].fillna(df[col].median())

# 3. Rellenamos las categóricas con 'Unknown'
for col in cols_categoricas:
    df[col] = df[col].fillna('Unknown')

print("✅ Valores nulos gestionados correctamente.")
print(f"Columnas numéricas tratadas: {len(cols_numericas)}")
print(f"Columnas de texto tratadas: {len(cols_categoricas)}")
```

```python
encoders = {}

# Lista de columnas de texto
categorical_cols = df.select_dtypes(include=['object']).columns

for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    encoders[col] = le
    print(f"✅ '{col}' — {len(le.classes_)} categorías: {list(le.classes_)}")

# ─── Guardar correctamente ────────────────────────────────────────────────────
import joblib, os
os.makedirs("../../models/Camila", exist_ok=True)

joblib.dump(encoders, '../../models/Camila/encoders.pkl')
print("\n💾 encoders.pkl guardado correctamente")
```

```python
# Guardamos la versión lista para Machine Learning
ruta_save = '../../data/processed/Camila/coffee_preprocessed.csv'
df.to_csv(ruta_save, index=False)

print(f"💾 Archivo guardado en: {ruta_save}")
```
